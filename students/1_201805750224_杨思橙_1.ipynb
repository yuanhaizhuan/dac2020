```python
import requests
import json
url_get='http://httpbin.org/get'
response=requests.get(url_get,timeout=5)
print(json.loads(response.text)['args'])
```

```python
from urllib import request,parse
url='http://httpbin.org/post'
headers={"user-agent":'mozilla/4.0(compatible;MSIE5.5;Windows NT)',
        "host":'httpbin.org'
        }
dict={"name":"hackdata"}
data=bytes(parse.urlencode(dict),encoding="utf8")
req=request.Request(url=url,data=data,headers=headers,method="post")
response=request.urlopen(req)
print(response.read().decode("utf-8"))
```

```python
import requests
url_cookies='http://httpbin.org/cookies'
url_set_cookies='http://httpbin.org/cookies/set?k1=v1&k2=v2'
print(requests.get(url_set_cookies,timeout=5).json())
print(requests.get(url_cookies,timeout=5).json())
```

```python
import requests
url_cookies='http://httpbin.org/cookies'
url_set_cookies='http://httpbin.org/cookies/set?k1=v1&k2=v2'
session=requests.session()
print(session.get(url_cookies,timeout=5).json())
print(session.get(url_set_cookies,timeout=5).json())
print(session.get(url_cookies,timeout=5).json())
```
