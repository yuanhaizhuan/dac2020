### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
# 这里编写代码
import urllib

'''
这里我模拟了必应搜索，搜索引擎用的就是get请求方法，一般网址变换出现？&就是get请求

下方的（search？q = 搜索关键字）是必应通用格式，可以自己探索

再举个例子，百度的（www.baidu.com/s?wd = 搜索关键字）
'''

url = "https://cn.bing.com/search"

word = {"q":"湘潭大学"}

word = urllib.parse.urlencode(word)                       #编码成字符串（中文一定要进行转码！！）

newurl = url + "?" + word                                 #拼接网址

request = urllib.request.Request(newurl)                  #发起请求

response = urllib.request.urlopen(request)                #打开页面

print(response.read().decode('utf-8'))                    #输出信息
```

基本上爬取的没问题，往下面翻翻就能看见湘潭大学的一些基本信息

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 深度遍历不去重算法
"""
深度遍历，这里我直接用了PPT上面的代码，就对如下做一些解释

正则表达式的Pattern对象是怎么出来的？
这里先说明个情况，进入开发者模式，然后左上角的小箭头点一点，放到要跳转的超链接上面，就会显示一串
例如<a href="https://github.com/requests/httpbin" class="github-corner" aria-label="在Github上查看源代码">
这里就通配的这么一个框架中的href，r'...'是为了防止里面翻译成转义字符，就显示原来的样子

为什么要用stack.pop(-1)？
其实就用到了切片编码原理，首为0的地方做栈底，后面后进先出

"""
import requests,re

# 设置层数
count = 10

# 将正则表达式编译成Pattern对象（r'原始字符串操作符）
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')

# 设置种子网站
seed = 'http://httpbin.org'

# 种子网站压栈
stack = [seed]

# 存储网站中url数量键值对
storage = {}


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; rv:2.0.1) Gecko/20200308 Firefox/4.0.1'}

while len(stack)>0 and count>0:
    try:
        url = stack.pop(-1)                                 # 末尾url网站出栈，后进先出
        html = requests.get(url,headers=headers).text       # 用requests库发出请求获取网页文本信息
        new_urls = r.findall(html)                          # 使用Pattern匹配网站文本,生成一个网站列表
        stack.extend(new_urls)                              # 子网站压栈
        storage[url] = len(new_urls)                        # 存储这层url下的子网站数量
        print(url+ "下的子网站数目为 " +"%d"%storage[url])
        count -= 1
    except Exception as e:
        print(url)                                          # 存储报错网站，方便寻找错误
        print(e)                                            # 存储报错原因
```

不去重的情况下，网站开始后无限循环爬取，如果没有设置count计数终止循环，后果不堪设想

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 广度遍历不去重算法
import requests,re

# 设置层数
count = 20

# 将正则表达式编译成Pattern对象（r'原始字符串操作符）
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')

# 设置种子网站
seed = 'http://www.baidu.com/'

# 种子网站进队
queue = [seed]

# 存储网站url数量键值对
storage = {}

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; rv:2.0.1) Gecko/20200308 Firefox/4.0.1'}

while len(queue)>0 and count>0:
    try:
        url = queue.pop(0)                                    # 退队，先进先出
        html = requests.get(url,headers = headers).text       # 用requests库发出请求获取网页文本信息
        new_urls = r.findall(html)                            # 使用Pattern匹配网站文本,生成一个网站列表
        queue.extend(new_urls)                                # 子网站进队
        storage[url] = len(new_urls)                          # 存储这层url下的子网站数量
        count -= 1
        print(url+ " 下的子网站数目为 " +"%d"%storage[url])
    except Exception as e:
        print(url)                                            # 存储报错网站，方便寻找错误
        print(e)                                               # 存储报错原因
```

可以看到有个passport.什么的网站重复爬了一次，这就很影响效率了

```python
# 广度遍历去重算法
import requests,re

# 设置层数
count = 20

# 将正则表达式编译成Pattern对象（r'原始字符串操作符）
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')

# 设置种子网站
seed = 'http://www.baidu.com'

# 种子网站进队
queue = [seed]

# 设置一个集合，保存已经抓取过的 URL
used = set()

# 存储网站url数量键值对
storage = {}

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; rv:2.0.1) Gecko/20200308 Firefox/4.0.1'}

while len(queue)>0 and count>0:
    try:
        url = queue.pop(0)                                    # 退队，先进先出
        used.add(url)                                         # 存放已经抓取过的url
        html = requests.get(url,headers = headers).text       # 用requests库发出请求获取网页文本信息
        new_urls = r.findall(html)                            # 使用Pattern匹配网站文本,生成一个网站列表
        for new_url in new_urls:
            if new_url not in used and new_url not in queue:
                queue.append(new_url)                          # 子网站进队
        storage[url] = len(new_urls)                           # 存储这层url下的子网站数量
        count -= 1
        print(url+ " 下的子网站数目为 " +"%d"%storage[url])
    except Exception as e:
        print(url)                                             # 存储报错网站，方便寻找错误
        print(e)                                               # 存储报错原因
```

和上面去重的一对比，显然passport什么的，没有重复爬，结果很OK！！
