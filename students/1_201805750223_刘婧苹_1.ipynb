### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
```

```python
import requests
import re
from collections import Counter
count=10
seed='https://www.baidu.com'
#response=requests.get(url=seed)
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
#headers={"User-Agent":'Mozilla/4.0(compatible;MSIE 5.5;Windows NT)',"host":'httpbin.org'}
#response=requests.get(url=seed,headers=headers)
stack=[seed]
storage={}
while len(stack)>0 and count>0:
    try:
        url=stack.pop(-1)
        html=requests.get(url).text
        #requests.adapters.DEFAULT_RETRIES = 5
        new_urls=r.findall(html)
        stack.extend(new_urls)
        storage[url]=len(new_urls)
        count-=1
    except Exception as e:
        print(url)
        print(e)
    
url_count=Counter(stack)
for url,count in url_count.most_common(10):
    print(url,count)
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
```

```python
#不去重
import requests
import re
from collections import Counter
count=10
seed='https://www.xtu.edu.cn/'
# response=requests.get(url=seed)
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
queue=[seed]
storage={}
while len(queue)>0 and count>0:
    try:
        url=queue.pop(0)
        html=requests.get(url).text
        new_urls=r.findall(html)
        queue.extend(new_urls)
        storage[url]=len(new_urls)
        count-=1
    except Exception as e:
        print(url)
        print(e)
        
url_count=Counter(queue)
for url,count in url_count.most_common(10):
    print(url,count)
```

```python
# 去重
import requests
import re
from collections import Counter
count=10
seed='https://www.baidu.com/'
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
queue=[seed]
storage={}
used=set()
while len(queue)>0 and count>0:
    try:
        url=queue.pop(0)
        html=requests.get(url).text
        storage[url]=html
        used.add(url)
        new_urls=r.findall(html)
        for new_url in new_urls:
            if new_url not in used and new_url not in queue:
                queue.extend(new_urls)
        count-=1
    except Exception as e:
        print(url)
        print(e)
        
        
url_count=Counter(queue)
for url,count in url_count.most_common(10):
    print(url,count)
```
